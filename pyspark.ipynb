```python
from pyspark.sql.functions import when

df_pyspark = df_pyspark.withColumn('NEW STATUS', when(df_pyspark.STATUS == 'Open', 1).otherwise(0)).sort('STATUS')
df_pyspark.show()
```

```python
df_pyspark = df_pyspark.withColumnRenamed('NEW STATUS', 'ANOTHER ONE')
df_pyspark.show()
```

```python
df_pyspark = df_pyspark.drop('ANOTHER ONE')
df_pyspark.show()
```
